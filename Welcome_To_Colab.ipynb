```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = 'API_SP.POP.TOTL_DS2_en_csv_v2_3401680.csv'  # Update with your file path
data = pd.read_csv(file_path, skiprows=4)  # Assuming the first few rows are metadata, adjust as needed

# Inspect the data to understand the structure
print("Data Info:")
print(data.info())
print("\nFirst few rows:")
print(data.head())

# Data Cleaning: Handle missing values if necessary
data.fillna(0, inplace=True)

# Plotting: Choose a column to plot (for example, population in a specific year)
# Replace 'Country Name' and a specific year like '2020' based on the actual column names
if 'Country Name' in data.columns and '2020' in data.columns:
    plt.figure(figsize=(12, 8))
    # Bar chart
    plt.bar(data['Country Name'], data['2020'], color='skyblue')
    plt.xticks(rotation=90)
    plt.title('Population in 2020 by Country')
    plt.xlabel('Country')
    plt.ylabel('Population')
    plt.tight_layout()
    plt.show()

    # Alternatively, plot a histogram of the population distribution for the year 2020
    plt.figure(figsize=(8, 6))
    sns.histplot(data['2020'], bins=30, kde=True)
    plt.title('Population Distribution in 2020')
    plt.xlabel('Population')
    plt.ylabel('Frequency')
    plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = 'Metadata_Indicator_API_SP.POP.TOTL_DS2_en_csv_v2_3401680.csv'  # Update with the correct file path
data = pd.read_csv(file_path)

# Inspect the data to understand the structure
print("Data Info:")
print(data.info())
print("\nFirst few rows:")
print(data.head())

# Data Cleaning: Handle missing values if necessary
data.fillna(0, inplace=True)

# Plotting: Example of plotting a bar chart or histogram if there are numeric columns
# Adjust the column names based on your data structure

# If the data contains country-related metadata and some numeric value, e.g., 'Indicator Name' and 'Some Numeric Column'
if 'Indicator Name' in data.columns and 'Some Numeric Column' in data.columns:
    plt.figure(figsize=(12, 8))
    # Bar chart
    plt.bar(data['Indicator Name'], data['Some Numeric Column'], color='skyblue')
    plt.xticks(rotation=90)
    plt.title('Bar Chart of Some Numeric Column by Indicator')
    plt.xlabel('Indicator Name')
    plt.ylabel('Numeric Value')
    plt.tight_layout()
    plt.show()

    # Alternatively, a histogram of some numeric column
    plt.figure(figsize=(8, 6))
    sns.histplot(data['Some Numeric Column'], bins=30, kde=True)
    plt.title('Histogram of Some Numeric Column')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()
```
